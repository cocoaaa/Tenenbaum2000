# How to use `Argparser` library to make a script (written in python language)
- Resource: https://dmorgan.info/posts/argparse-intro/


```python
from argparse import ArgumentParser
from pprint import pprint

parser = ArgumentParser()

# Add flags
parser.add_argument('--cities', nargs="+", default=["la", "paris"])
parser.add_argument('--styles', nargs="+", default=["CartoVoyagerNoLabels", "StamenTonerBackground"])
parser.add_argument('--zooms', nargs="+", default=["14"])
```

## Parse the system args (default) or a list of strings
- e.g. `parser.parse_args(["--cities", "seoul", "--styles" "Watercolor","Disney"])`
- The parsing rule is defined above (via. previous 'add_argument' statements)

See the implementation snippet of 'ArgumentParser.parse_known_args'

```python
# Parse
# args, unknown = parser.parse_known_args("--cities seoul --styles Watercolor".split())
args, unknown = parser.parse_known_args("--cities seoul chicago --styles Watercolor".split())

args
```

The returned output of 'parser.parse_args' is an `Namespace` object. It's quite like a `dict`. 
- Set an attribute of the Namepspace object `[keyname]` = `value` assignment.


```python
args['temp'] = 10
```
